```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
merge_df = pd.merge(school_data, student_data, on="school_name")
merge_df
```

```python
#calculate the total number of schools
total_schools = len(merge_df["school_name"].value_counts())
total_schools
```

```python
#calculate the total number of students
total_students = len(merge_df["Student ID"].value_counts())
total_students
```

```python
#calculate the total budget

budget = merge_df["budget"].unique()

total_budget = budget.sum()

total_budget
```

```python
#Calculate the average math score

avg_math = merge_df["math_score"].mean()
avg_math
```

```python
#Calculate the average reading score

avg_reading = merge_df ["reading_score"].mean()
avg_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

math_passing = merge_df.loc[merge_df["math_score"] >= 70]["math_score"].count()

percent_pass_math = (math_passing / total_students) * 100
percent_pass_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)


reading_passing = merge_df.loc[merge_df["reading_score"] >= 70]["reading_score"].count()

percent_pass_reading = reading_passing / total_students * 100
percent_pass_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)

total_pass = merge_df.loc[(merge_df["math_score"] >= 70) & (merge_df["reading_score"] >=70)]["student_name"].count()

percent_passing = total_pass / total_students * 100
percent_passing
```

```python
#Create a dataframe to hold the above results

district_summary_df = pd.DataFrame({
    
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math":[percent_pass_math],
    "% Passing Reading": [percent_pass_reading],
    "% Overall Passing": [percent_passing]

})

district_summary_df
```

```python
summary_styling_df = district_summary_df.style.format({"Total Budget": "${:,.2f}", 
                                                       "Total Students": "{:,}"})


summary_styling_df
```
